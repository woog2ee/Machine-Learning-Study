## import library

```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.mplot3d import Axes3D
```

## load point data for training and testing

```python
filename    = 'Assignment 06 data.csv'
data        = np.genfromtxt(filename, delimiter=',')
num_data    = data.shape[0]

xx = np.zeros(num_data)
yy = np.zeros(num_data)
zz = np.zeros(num_data)

for i in range(num_data):
    xx[i] = data[i,0]
    yy[i] = data[i,1]
    zz[i] = data[i,2]
```

## plot the data in the three dimensional space

```python
fig = plt.figure(figsize=(12,8))
ax  = plt.subplot(111, projection='3d')
plt.title('data')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')

ax.scatter(xx, yy, zz, marker='o', color='blue', alpha=0.2)

plt.tight_layout()
plt.show()
```

## compute the loss function

```python
def compute_residual(theta, x, y, z):
    f = np.zeros(num_data)
    for i in range(num_data):
        param = np.array((1, x[i], y[i])).reshape(3, 1)
        f[i] = sum(theta * param)
    
    residual = z - f
    return residual, num_data
```

```python
def compute_loss(theta, x, y, z):
    residual, num_data = compute_residual(theta, x, y, z)
    
    loss = 0
    for i in range(num_data):
        loss += np.power(residual[i], 2) / (2 * num_data)
        
    return loss
```

## compute the gradient for each model parameter

```python
def compute_gradient(theta, x, y, z):
    theta = theta.reshape(3, 1)
    Y = z.reshape(num_data, 1)
    
    X = np.zeros((num_data, 3))
    X[:, 0] = 1
    X[:, 1] = x[:]
    X[:, 2] = y[:]
    
    grad = (2 / num_data) * X.T @ (X @ theta - Y)
    return grad
```

## gradient descent for each model parameter

```python
num_iteration   = 1000
learning_rate   = 0.01

theta           = np.array((0, 0, 0)).reshape(3, 1)
theta_iteration = np.zeros((num_iteration, theta.size))
loss_iteration  = np.zeros(num_iteration)
```

```python
for i in range(num_iteration):
    theta  = theta - learning_rate * compute_gradient(theta, xx, yy, zz)
    loss   = compute_loss(theta, xx, yy, zz)
   
    theta_iteration[i] = theta.reshape(1, 3)
    loss_iteration[i]  = loss
    
    print("iteration = %4d, loss = %5.5f" % (i, loss))
```

```python
f = theta[0] + theta[1] * xx + theta[2] * yy
```

## plot the results

```python
XX, YY = np.meshgrid(xx, yy)
ZZ = theta[0] + theta[1]*XX + theta[2]*YY
```

```python
def plot_data(xx, yy, zz):
    fig = plt.figure(figsize=(12, 8))
    ax  = plt.subplot(111, projection='3d')
    plt.title('data')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')

    ax.scatter(xx, yy, zz, marker='o', color='blue', alpha=0.2)

    plt.tight_layout()
    plt.show()
    
    
    
def plot_model_parameter(theta_iteration):
    plt.figure(figsize=(8, 6))
    plt.title('model parameter')
    
    plt.plot(theta_iteration[:, 0], '-', color='red')
    plt.plot(theta_iteration[:, 1], '-', color='green')
    plt.plot(theta_iteration[:, 2], '-', color='blue')
    
    plt.tight_layout()
    plt.show()
    
    
    
def plot_loss_curve(loss_iteration):
    plt.figure(figsize=(8, 6))
    plt.title('loss')

    plt.plot(loss_iteration, '-', color='red')

    plt.tight_layout()
    plt.show()
    
    
    
def plot_surface(XX, YY, ZZ, xx, yy, zz):
    fig = plt.figure(figsize=(12, 8))
    ax  = plt.subplot(111, projection='3d')
    plt.title('optimal regression surface')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')

    ax.scatter(xx, yy, zz, marker='o', color='blue', alpha=0.2)
    ax.plot_surface(XX, YY, ZZ, cmap='Greys')
    
    plt.tight_layout()
    plt.show()
```
